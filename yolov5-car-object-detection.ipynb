# **Visualize Predictions**

```python
import os
file_paths = []
flist = os.listdir('home/zahid/Downloads/objectDetection ssd/data/exp')
for file in flist:
    fpath = os.path.join('home/zahid/Downloads/objectDetection ssd/data/exp', file)
    file_paths.append(fpath)
```

```python
import numpy as np
random = np.random.randint(0, len(file_paths) - 1)
img_path = file_paths[random]
print(img_path)
```

```python
plt.figure(figsize=(30, 25))
for i in range(25):
    random = np.random.randint(0, len(file_paths) - 1)
    plt.subplot(5, 5, i + 1)
    img_path = file_paths[random]
    img = plt.imread(img_path)
    plt.imshow(img)
    plt.title('Predictions', size=10, color="black") 
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

As we can see, yolov5 large is great at car detection and also picked up some traffic lights and a stop sign.
